```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

def model(w, b, x):
    return 1 if np.dot(w, x) >= b else 0

def predict(w, b, X):
    return np.array([model(w, b, x) for x in X])

def fit(X, y, epochs=1, lr=1):
    # Initialize weights and bias
    w = np.ones(X.shape[1])
    b = 0
    
    
    accuracy = []
    max_accuracy = 0
    wt_matrix = []
    
    for i in range(epochs):
        for x_i, y_i in zip(X, y):  
            y_pred = model(w, b, x_i)
            if y_i == 1 and y_pred == 0:
                w += lr * x_i
                b += lr
            elif y_i == 0 and y_pred == 1:
                w -= lr * x_i
                b -= lr

            wt_matrix.append(w.copy())
        
       
        y_pred_all = predict(w, b, X)
        epoch_accuracy = accuracy_score(y, y_pred_all)
        accuracy.append(epoch_accuracy)
        
        if epoch_accuracy > max_accuracy:
            max_accuracy = epoch_accuracy
            chkptw = w.copy()
            chkptb = b
        
      
        w, b = chkptw.copy(), chkptb
        
        print(f"Epoch {i + 1}: Max Accuracy: {max_accuracy:.4f}")
    
    plt.plot(range(epochs), accuracy)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.ylim([0, 1])
    plt.title('Accuracy over Epochs')
    plt.show()
    
    return np.array(wt_matrix), w, b

```
